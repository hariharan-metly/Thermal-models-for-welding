# Rosenthal 2D solution

$$T=T_i + \frac{\eta P}{2\pi k g}\exp\left(\frac{Vx}{2\alpha}\right)K_0\left(\frac{Vr}{2\alpha}\right)$$

$K_0$ - Zeroth order Modified Bessel function of second kind - Solution to Modified Bessel Differential equation

## Modified Bessel differential equation

$$z^2\frac{d^2y}{dz^2}+z\frac{dy}{dz}-(z^2+v^2)y=0$$

$v$ Order <br>

$z$ Argument <br>

```python
#Import the necessary packages

import numpy as np
from scipy import special
import matplotlib.pyplot as plt
```

```python
# Visualize the Modified Bessel function of second kind and order zero
z=np.linspace(0,1)
plt.plot(z,special.kv(0,z),linewidth='3')
plt.xlabel('x')
plt.ylabel('Bessel function')
plt.gcf().set_dpi(300)
```

```python
#Data and setup
a=8.5e-6
g=2E-3
k=35
Ti=298
v=2e-3
P=1000
```

```python
x0,y0=0.0,0.0
```

```python
x=np.linspace(-5E-2,5E-2,100)
y=np.linspace(-5E-2,5E-2,100)
X,Y=np.meshgrid(x,y)
R=np.sqrt((X-x0)**2+(Y-y0)**2)
```

```python
T=Ti+(P/(2*np.pi*k*g))*np.exp(v*X/(2*a))*special.kv(0,(v*R/(2*a)))
```

```python
cs = plt.contourf(X,Y,T,levels=np.arange(298,1800), cmap=plt.cm.coolwarm, extend="both")
cs.set_clim(298, 1800)
plt.gcf().set_dpi(300)
cb = plt.colorbar(cs)
```
